```
# Define the URL of the website you want to analyze
url = "https://www.werrv.ca/"
# Use the requests library to fetch the HTML of the website
response = requests.get(url)
html = response.text
# Use BeautifulSoup to parse the HTML
soup = BeautifulSoup(html, "html.parser")
# Check the title tag of the website
title_tag = soup.find("title")
title = title_tag.text
print("Title:", title)
# Check the description tag of the website
description_tag = soup.find("meta", attrs={"name": "description"})
description = description_tag["content"]
print("Description:", description)
# Check the number of internal links on the website
internal_links = []
for link in soup.find_all("a"):
    href = link.get("href")
    if href.startswith("/") and href not in internal_links:
        internal_links.append(href)
print("Number of internal links:", len(internal_links))
```

```
# html tag cleaner
CLEANR = re.compile('<.*?>') 

def cleanhtml(raw_html):
  cleantext = re.sub(CLEANR, '', raw_html)
  return cleantext

Body_cleaned=[]
for des in df['Body HTML']:
  text=cleanhtml(des)
  Body_cleaned=np.append(Body_cleaned,text)
```

```
df['Body_cleaned']=Body_cleaned
```

```
df.head(10)
```

```
b_model = Summarizer() 
```

```
meta_description=[]

for description in df['Body_cleaned']:
  text = b_model(description,min_length= 160,ratio=0.01)
  meta_description = np.append(meta_description,text)
```

```
df['Meta_Description_raw'] = meta_description
```
